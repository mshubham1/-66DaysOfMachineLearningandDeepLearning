# **Seaborn**
Seaborn is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.
* Distribution plots
* Distplot
* Joinplot
* Pairplot
* COUNTPLOT

```python
import seaborn as sns 
data = sns.load_dataset("iris") 
data.head()
```

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
#relplot
import matplotlib.pyplot as plt
sns.relplot(x='sepal_width', y='species', data=data)
plt.show()
```

```python
## Correlation with Heatmap
#A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
#JoinPlot
#A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
sns.jointplot(x='petal_length',y='petal_width',data=data,kind='hex')
```

```python
# PAIR PLOT
#A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables
sns.pairplot(data)
```

```python
sns.pairplot(data,hue='species')
```

```python
#DIST PLOT
#Dist plot helps us to check the distribution of the columns feature

sns.distplot(data['petal_length'])
```

```python
sns.distplot(data['petal_length'],kde=False,bins=10)
```

**Categorical Plots**
Seaborn also helps us in doing the analysis on Categorical Data points.
* boxplot
* violinplot
* countplot
* bar plot





```python
#Count plot - helps to find the count of the category
sns.countplot('species',data=data)
```

```python
#barpolot
sns.barplot(x='species', y='sepal_length', data=data)
```

```python
#boxplot - It shows the distribution of the quantitative data that represents the comparisons between variables. 
#boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution
sns.boxplot(x='species', y='sepal_width', data=data)
```

```python
#vinplot - t is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.
sns.violinplot(x='species', y='sepal_width', data=data)
```

```python
# KDE -KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.
# It depicts the probability density at different values in a continuous variable. 
#We can also plot a single graph for multiple samples which helps in more efficient data visualization.
```

```python
sns.kdeplot(x='sepal_length', y='sepal_width', data=data)
```
